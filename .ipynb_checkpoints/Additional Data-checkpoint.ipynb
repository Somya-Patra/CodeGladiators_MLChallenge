## Model application

```python
import pandas as pd
import numpy as np
```

```python
scal_ex = pd.read_csv("files/FINAL_TRAINING_DATA.csv")
label = scal_ex["conversion_fraud"]
scal_ex = scal_ex.drop(["conversion_fraud"],axis=1)

scal_ex
```

```python
scal_ex= scal_ex.drop(['clmbuserid_cr','imprid_cr'],axis=1)
```

```python
scal_ex
```

```python
from sklearn.ensemble import RandomForestClassifier
#from sklearn.tree import DecisionTreeClassifier
#from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(scal_ex, label, test_size=0.2)
```

```python
box = [10,20,30,40,50,60,70,100,200]
```

```python
from tqdm import tqdm
l = []
for i in tqdm(range(len(box)),desc="Progress"):
    model  = RandomForestClassifier(random_state=42,n_estimators=box[i])
    model.fit(X_train,y_train)
    
    TR_pred = model.predict(X_train)
    VA_pred = model.predict(X_val)
    
    TR_acc = accuracy_score(TR_pred,y_train)
    VA_acc = accuracy_score(VA_pred,y_val)
    
    ovrfit = TR_acc-VA_acc
    
    l.append([box[i],TR_acc,VA_acc,ovrfit])
```

```python
l
```

## Gridsearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
from xgboost import XGBClassifier
```

```python
#rf = RandomForestClassifier(random_state=42)
xg = XGBClassifier(objective="binary:logistic",eval_metric="auc")
```

```python
'''  rf
params = {'bootstrap': [True, False],
 'max_depth': [10, 20, 30, 40, 50],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [20, 40, 60, 80, 100]}
'''

params = {
        'min_child_weight': [0.8,1,2],
        'gamma': [0.02,0.04,0.07,0.01,0.2],
        'subsample': [0.03,0.008,0.01,0.02],
        'colsample_bytree': [0.1],
        'max_depth': [2]
        
        }

#params = {'gamma': [0.01,0.02,0.03,0.07,0.1,0.2,0.5],}
```

```python
grid = GridSearchCV(estimator = xg, param_grid = params, cv = 8, verbose=10)
grid_result = grid.fit(scal_ex,label)
```

```python
grid_result.best_score_
```
